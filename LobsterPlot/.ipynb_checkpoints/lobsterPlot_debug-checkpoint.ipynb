## Here's where we load the produced files and define the plot limts

```python
########################################################################### 
###  best fit values,       best fit - 3sigma,   best fit + 3sigma      ###
########################################################################### 

T12 = [arcsin(sqrt(0.304)), arcsin(sqrt(0.265)), arcsin(sqrt(0.346))]
T13_N = [arcsin(sqrt(0.0219)), arcsin(sqrt(0.0183)), arcsin(sqrt(0.0255))]
T13_I = [arcsin(sqrt(0.0219)), arcsin(sqrt(0.0183)), arcsin(sqrt(0.0255))] 
## all in meV**2
dM2_12 = [7.53E1, 6.99E1, 8.07E1]
dM2_N = [2.42E3, 2.24E3, 2.60E3]
dM2_13_N = [i+0.5*j for i,j in zip(dM2_N, dM2_12)]
dM2_I = [2.49E3, 2.31E3, 2.67E3]
dM2_13_I = [i-0.5*j for i,j in zip(dM2_I, dM2_12)]

########### STERILE Neutrino Mixing ##########
T14 = [arcsin(sqrt(0.09))/2, arcsin(sqrt(0.09))/2, arcsin(sqrt(0.09))/2]
dM2_14 = [1.78E6,1.78E6,1.78E6]
```

```python
ml = ml = np.linspace(0.01,1,100)
ml = np.append(ml, np.linspace(1.1,10,90))
ml = np.append(ml, np.linspace(11,100,90))
ml = np.append(ml, np.linspace(110,1000,90))
N =len(ml)
```

```python

mbbN = np.zeros((N,4)) 
mbbI = np.zeros((N,4))
mbN = np.zeros((N,3))
mbI = np.zeros((N,3))
sumMN = np.zeros((N,3))
sumMI = np.zeros((N,3))

for i in range (len(ml)):
    mbbN[i], mbN[i], sumMN[i] = \
            nuM_range(ml[i], 'NH', False, [T12, dM2_12], [T13_N, dM2_13_N],
            [T14, dM2_14], 100)
    mbbI[i], mbI[i], sumMI[i] = \
            nuM_range(ml[i], 'IH', False, [T12, dM2_12], [T13_I, dM2_13_I],
            [T14, dM2_14], 100)
```

```python
def AddExperimentalLimits(IH, NH, xMin, xMax, isotopes=None, yMin=-1, yMax=-1):
    """
    This function adds specified limits and sensitivities to Lobster plot

    Parameters
    ----------
    ax : axes object
        matplotlib object where plot will be drawn.

    xArray : array_like
        This is the x-axis (energy in meV) array.

    y4DArray : array_like
        This is a 4D array containing the central value spread (min, max, respectively) in the first two internal arrays y4DArray[:,1], y4DArray[:,2], and the 3sigma spread values (min, max, respectively) in y4DArray[:,2], y4DArray[:,3].

    col : string
        choose a color based on matplotlib color palettes: https://matplotlib.org/stable/gallery/color/named_colors.html

    xlab : string
        x-axis label

    ylab : string
        y-axis label

    Returns
    -------
    plot object
        It returns a plot object.
    """

    arrowXscale = 1.2
    axSpan = xMax-xMin
    if(isotopes==None):
        isotopes = ["Xe", "Te", "Ge"]
        
    ##Te limit from CUORE
    
    if("Te" in isotopes):
        mbb_min_Te = 90
        mbb_max_Te = 305
        A_Te = 130
        IH.axhspan(mbb_min_Te, mbb_max_Te, xmin = (A_Te)/axSpan, xmax = (A_Te+20)/axSpan, lw=0,fc='#2B4970',ec='#2B4970', alpha=0.3)
        Teline = IH.hlines(mbb_min_Te, A_Te-5, A_Te+20, color='#2B4970', label='$^{130}$Te limit (CUORE [Prelim.])')
        IH.errorbar(A_Te, mbb_min_Te,yerr = mbb_max_Te-mbb_min_Te-20, lolims=True,  color='#2B4970')
        IH.text(A_Te, mbb_min_Te, '$^{130}$Te', color='#2B4970',fontsize='small')

    ## Ge limit from GERDA (2020): https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.125.252502
    if("Ge" in isotopes):
        mbb_min_Ge = 79
        mbb_max_Ge = 180
        A_Ge = 76
        IH.axhspan(mbb_min_Ge, mbb_max_Ge, xmin = (A_Ge)/axSpan, xmax = (A_Ge+20)/axSpan, lw=0, ec='#AA7F39',fill=None, alpha=0.3, hatch='\\\\\\')
        Geline = IH.hlines(mbb_min_Ge, A_Ge-5, A_Ge+20, color='#AA7F39', label='$^{76}$Ge limit (GERDA)', linestyle='-')
        IH.errorbar(A_Ge, mbb_min_Ge, yerr=mbb_max_Ge-mbb_min_Ge-20, lolims=True,  color='#AA7F39')

    ##Xe limit from KamLand-Zen (2016): https://journals.aps.org/prl/pdf/10.1103/PhysRevLett.117.082503
    if("Xe" in isotopes):
        mbb_min_Xe = 61
        mbb_max_Xe = 165
        A_Xe = 136
        
        IH.axhspan(mbb_min_Xe, mbb_max_Xe, xmin = (A_Xe)/axSpan, xmax = (A_Xe+20)/axSpan, lw=0, ec='#AA7F39',fill=None, alpha=0.3, hatch='///')
        Xeline = IH.hlines(mbb_min_Xe, A_Xe-5, A_Xe+20, color='#AA7F39', label='$^{136}$Xe limit (KamLAND-Zen)', linestyle='-')
        IH.errorbar(A_Xe, mbb_min_Xe, yerr=mbb_max_Xe-mbb_min_Xe-20, lolims=True,  color='#AA7F39')
    
    ## Mo limit from CUPID-Mo (2021): https://journals.aps.org/prl/abstract/10.1103/PhysRevLett.126.1
    if("Mo" in isotopes):
        mbb_min_Mo = 300
        mbb_max_Mo = 500
        A_Mo = 100
        
        IH.axhspan(300, 500,lw=0, ec='#AA7F39',fill=None, alpha=0.3, hatch='///')
        Xeline = IH.axhline(300, color='#AA7F39', label='$^{100}$Mo limit (CUPID-Mo)', linestyle='-')
        IH.errorbar(xMin*6*arrowXscale, 120, yerr=120, lolims=True,  color='#AA7F39')
    

    ##
    ## Sensitivities ##
    ##
    
#     IH.axhspan(50, 130,lw=0, fc='#2B4970',ec='#2B4970', alpha=0.3)
#     IH.axhline(50, color='#2B4970', label='CUORE Sensitivity')
#     IH.errorbar(xMin*arrowXscale, 50, yerr=50, lolims=True,  color='#2B4970')
#     IH.text(xMin*pow(arrowXscale,2), 50*1.2, 'CUORE Sensitivity', color='#2B4970',fontsize='small')

#     IH.legend(handles=[Teline, Geline, Xeline], loc=3,prop={'size':9})

    # ON the right panel
#     NH.axhspan(270, 650, lw=0,fc='#2B4970',ec='#AA9B39', alpha=0.3, label='Te-130 Limit')
#     #NH.axhspan(270, 650, lw=0,ec='#AA9B39',fill=None, alpha=0.3, hatch='///')
#     NH.axhline(270, color='#2B4970', label='Te-130 Limit')

#     NH.axhspan(200, 400, lw=0,ec='#AA7F39',fill=None, alpha=0.3, hatch='\\\\\\')
#     NH.axhline(200, color='#AA7F39', label='Ge-76 Limit', linestyle='--')

#     NH.axhspan(120, 250,lw=0, ec='#AA7F39',fill=None, alpha=0.3, hatch='///')
#     NH.axhline(120, color='#AA7F39', label='Xe-136 Limit', linestyle='-')

#     NH.axhspan(50, 130, lw=0,fc='#2B4970',ec='#452F74', alpha=0.3, label='CUORE Sensitivity')
#     #NH.axhspan(50, 130,lw=0, ec='#452F74',fill=None, alpha=0.3, hatch='\\\\\\')
#     NH.axhline(50, color='#2B4970', label='CUORE Sensitivity')

    # IH.set_title('3 Flavors, Inverted Hierarchy')
    # NH.set_title('3 Flavors, Normal Hierarchy')

    IH.set_xlim(xMin, xMax)
    NH.set_xlim(xMin, xMax)

    if yMin>0 and yMax >0:
        IH.set_ylim(yMin, yMax)
    else:
        IH.set_ylim(xMin, xMax)
```

```python
plt.xkcd()  # Yes...

# plt.plot(sin(linspace(0, 10)))
# plt.title('Whoo Hoo!!!')

f, (IH, NH) = plt.subplots(1, 2, sharex=False, sharey=True, figsize=(8,6),gridspec_kw={'width_ratios': [2.5, 1]})
# with plt.xkcd():
massContour(IH, ml, mbbI, 'C0',r'$m_{3}$ (meV)', r'$m_{\beta\beta}$ (meV)')
massContour(IH, ml, mbbN, 'C3',r'$m_{lightest}$ (meV)',r'$m_{\beta\beta}$ (meV)')

f.subplots_adjust(wspace=0, left = 0.12, bottom=0.15, right=0.97, top=0.95)

# bestfit = mpatches.Patch(color='g', alpha=0.4)
# threeS = mpatches.Patch(color='g', alpha=0.2)
# IH.legend([bestfit, threeS], ['Best fit', '3$\sigma$ band'],loc=2,prop={'size':11},title='Inverted Hierarchy')
# bestfit = mpatches.Patch(color='r', alpha=0.4)
# threeS = mpatches.Patch(color='r', alpha=0.2)
# NH.legend([bestfit, threeS], ['Best fit', '3$\sigma$ band'],loc=2,prop={'size':11},title='Normal Hierarchy')
AddExperimentalLimits(NH, NH, xMin=70, xMax=200, isotopes=["Te", "Ge", "Xe", ])

IH.set_xlim(.2,500)
IH.set_ylim(.2,500)
# IH.grid()
```
